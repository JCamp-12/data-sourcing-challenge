### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
import urllib.parse
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

### Access the New York Times API

```python
# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name%3A%22Movies%22%20AND%20type_of_material%3A%22Review%22%20AND%20headline%3A%22love%22'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL

query_url = (f"{url}api-key={nyt_api_key}&begin_date={begin_date}&end_date={end_date}&fq={filter_query}&sort={sort}&fl={field_list}")
query_url

# movies = requests.get(query_url).json()

# movie_list = movies["response"]["docs"]
# print(json.dumps(movie_list, indent=4))
```

```python
# Create an empty list to store the reviews
review_list = []

# loop through pages 0-19
for page in range(0, 20):
# for page in range(0, 2):   
    
    # create query with a page number
    # API results show 10 articles at a time
    print('page', page)
    final_query_url = f"{query_url}&page={str(page)}"
    print(final_query_url)
    
    # Make a "GET" request and retrieve the JSON
    movies = requests.get(final_query_url).json()
    movie_list = movies["response"]["docs"]
    
    # first_five = movie_list[:3]
    # print(json.dumps(first_five, indent=4))
    
    # Add a twelve second interval between queries to stay within API query limits
    time.sleep(12)
    
    # Try and save the reviews to the reviews_list
    try:
        for movie in movie_list:
        # loop through the reviews["response"]["docs"] and append each review to the list
            review_list.append(movie)
        # Print the page that was just retrieved
        print(f"Checked page {page}")
        # Print the page number that had no results then break from the loop
    except:
        print(f"No results. Ended at page {page}.")
        break
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
length_check = len(review_list)
print('length check', length_check)
first_five = review_list[:5]
print(json.dumps(first_five, indent=4))
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
review_list_df = pd.json_normalize(review_list)
review_list_df.head(20)
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early

review_list_df['title'] = review_list_df['headline.main'].apply(lambda st: st[st.find("\u2018")+1:st.find("\u2019 Review")])
review_list_df
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
review_list_df['keywords'] = review_list_df['keywords'].apply(lambda x: extract_keywords(x))
review_list_df.head(20)
```

```python
# Create a list from the "title" column using to_list()
# These titles will be used in the query for The Movie Database
title_list = review_list_df['title'].tolist()
print(title_list)
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key
```

```python
# Create an empty list to store the results
tmdb_movies_list = []



# Create a request counter to sleep the requests after a multiple
# of 50 requests
request_counter = 1

# Loop through the titles
print('length of titles', len(title_list))
for title in title_list:
    # Check if we need to sleep before making a request
    if request_counter == 50:
        print('application is sleeping')
        request_counter = 1
        time.sleep(1)
    # Add 1 to the request counter
    else:
        request_counter = request_counter + 1
    
    # Perform a "GET" request for The Movie Database
    # "https://api.themoviedb.org/3/search/movie?query=the%20godfather&include_adult=false&language=en-US&page=1"
    encoded_title = urllib.parse.quote(title)
    final_url = url + encoded_title + tmdb_key_string
    # print('url', final_url)
    response_data = requests.get(final_url)
    data = response_data.json()
    
    # Use json.dumps with argument indent=4 to format data
    # print(json.dumps(data, indent=4))
    
    # Include a try clause to search for the full movie details.
    # Use the except clause to print out a statement if a movie
    # is not found.
    try:
        # Get movie id
        movie_id = data["results"][0]['id']
        movie_id = str(movie_id)
        # print('movie id ', movie_id)
        
        # Make a request for a the full movie details
        
        movie_detail_url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={tmdb_api_key}"
        # f"https://api.themoviedb.org/3/movie/{movie_id}{tmdb_key_string}"
        
        # print(movie_detail_url)
        # Execute "GET" request with url
        detail_response_data = requests.get(movie_detail_url)
        detail_response_json = detail_response_data.json()
        if request_counter == 2:
            print(json.dumps(detail_response_json, indent=4))
        
        # Extract the genre names into a list
        for genre_obj in detail_response_json["genres"]:
            genres = []
            genres.append(genre_obj["name"])
        
        # Extract the spoken_languages' English name into a list
        for spoken_obj in detail_response_json["spoken_languages"]:
            spoken_languages = []
            spoken_languages.append(spoken_obj["english_name"])
        # Extract the production_countries' name into a list
        for production_countries_obj in detail_response_json["production_countries"]:
            production_countries = []
            production_countries.append(production_countries_obj["name"])
        # Add the relevant data to a dictionary and
        # append it to the tmdb_movies_list list
        tmdb_movies_dict = {
            "title": detail_response_json["title"],
            "original_title": detail_response_json["original_title"],
            "budget": detail_response_json["budget"],
            "original_language": detail_response_json["original_language"],
            "homepage": detail_response_json["homepage"],
            "overview": detail_response_json["overview"],
            "popularity": detail_response_json["popularity"],
            "runtime": detail_response_json["runtime"],
            "revenue": detail_response_json["revenue"],
            "release_date": detail_response_json["release_date"],
            "vote_average": detail_response_json["vote_average"],
            "vote_count": detail_response_json["vote_count"],
            "production_countries": production_countries,
            "spoken_languages": spoken_languages,
            "genres": genres
            
        }
        tmdb_movies_list.append(tmdb_movies_dict)
        # title, original_title, budget, original_language, homepage, overview, popularity, 
        # runtime, revenue, release_date, vote_average, vote_count,
        # as well as the genres, spoken_languages, and production_countries lists you just created.

        
        
        # Print out the title that was found
        print("Movie Found: ", title)
    except:
        print("Movie Not Found: ", title)


# print('tmdb_movies_list: ', tmdb_movies_list)
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data

first_five_movie_list = json.dumps(tmdb_movies_list[:5], indent=4)
print(first_five_movie_list)
```

```python
# Convert the results to a DataFrame
tmdb_df = pd.DataFrame(tmdb_movies_list)
tmdb_df
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title
merge_movie_df = pd.merge(review_list_df, tmdb_df, on="title")
merge_movie_df
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
columns_to_fix = ["production_countries", "spoken_languages", "genres"]

# Create a list of characters to remove
characters_to_remove = ["[", "]", "'"]

# Loop through the list of columns to fix
for column in columns_to_fix:
    # Convert the column to type 'str'
    merge_movie_df[column] = merge_movie_df.loc[:, column].astype(str)

    # Loop through characters to remove
    for character in characters_to_remove:
        merge_movie_df[column] = merge_movie_df[column].str.replace(character, '')
# Display the fixed DataFrame
merge_movie_df
```

```python
# Drop "byline.person" column
merge_movie_df.drop("byline.person", axis=1, inplace=True)
merge_movie_df.drop("web_url", axis=1, inplace=True)
merge_movie_df
```

```python
# Delete duplicate rows and reset index
merge_movie_df = merge_movie_df.drop_duplicates()
merge_movie_df = merge_movie_df.reset_index(drop=True)
merge_movie_df
```

```python
# Export data to CSV without the index
merge_movie_df.to_csv("output/final_movies.csv", index=False)
```
